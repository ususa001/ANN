````python
import requests
response = requests.get("https://raw.githubusercontent.com/gaith7/Data-Engineering-Course/main/Datasets/input_text.txt")
data = response.text
text=[]
for i, line in enumerate(data.split('\n')):
    print(f'{i}   {line}')
    text.append(line)
````

````python
# save text in a file

file = open("myfile.txt","w")

for line in text:
  file.writelines(line+"\n")


file.close()
````

````python
# read a text file
Text=[]
data = open("myfile.txt", "r")

while(True):
	#read next line
	line = data.readline()
	#if line is empty, you are done with all lines in the file
	if not line:
		break
	#you can access the line

	Text.append(line.strip())
	print(line.strip())
````
